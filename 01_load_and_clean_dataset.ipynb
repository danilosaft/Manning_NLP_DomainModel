```python
def removeHTML(item):
    pattern = r"<[^>]*>"
    regexp = re.compile(pattern)
    item = regexp.sub("", item)
    return item
    
def removeLatex(item):
    pattern = r"\$*\$"
    regexp = re.compile(pattern)
    item = regexp.sub("", item)
    return item

def removeURLs(item):
    pattern = r"https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)"
    regexp = re.compile(pattern)
    item = regexp.sub("", item)
    pattern = r"[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)"
    regexp = re.compile(pattern)
    item = regexp.sub("", item)
    #pattern = r"www.*\s"
    #regexp = re.compile(pattern)
    #item = regexp.sub("", item)
    #pattern = r"\s*.[a-zA-Z\d]{2,3}\s"
    #regexp = re.compile(pattern)
    #item = regexp.sub("", item)
    #pattern = r"\s*.[a-zA-Z\d]{2,3}\/*\s"
    #regexp = re.compile(pattern)
    #item = regexp.sub("", item)
    return item

def removeDigitsLineReturnsAndSoOn(item):
    pattern = r"%d|\n|\t|\r"
    regexp = re.compile(pattern)
    item = regexp.sub("", item)
    return item


df['text'] = df['text'].map(removeHTML)
df['text'] = df['text'].map(removeLatex)
df['text'] = df['text'].map(removeURLs)
#df['text'] = df['text'].map(removeDigitsLineReturnsAndSoOn)

df.groupby("category").head(20)
```

```python
df["text"] = df["text"].str.strip()
df["text"].replace('', np.nan, inplace=True)
df.dropna(subset=["text"], inplace = True)
df.shape
```

```python
df.head()
```

```python
def countWords(item):
    matches = re.findall("[\\w-]+",item)
    return len(matches)   

df = df[df["text"].map(countWords) > 5]
df = df[df["text"].map(countWords) < 500]
df.shape
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
df['tokenized_text'] = df['text'].apply(word_tokenize) 
df.shape
```
